# Country of birth for people granted New Zealand Citizenship

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('granted-citizenship-1949-2020.csv')
df
```

```python
df.describe()
```

```python
df.columns
```

```python
df = df.drop(labels=334, axis=0)
df
```

```python
df['1949'] = df.iloc[:,3:4].sum(axis=1)
df['1950-1959'] = df.iloc[:,4:14].sum(axis=1)
df['1960-1069'] = df.iloc[:,14:24].sum(axis=1)
df['1970-1979'] = df.iloc[:,24:34].sum(axis=1)
df['1980-1989'] = df.iloc[:,34:44].sum(axis=1)
df['1990-1999'] = df.iloc[:,44:54].sum(axis=1)
df['2000-2009'] = df.iloc[:,54:64].sum(axis=1)
df['2010-2020'] = df.iloc[:,64:75].sum(axis=1)
```

```python
years = ['1949','1950-1959','1960-1069','1970-1979','1980-1989','1990-1999','2000-2009','2010-2020']
df.sort_values(['Total'], ascending=False, axis=0, inplace=True)
df_top5 = df.head()
df_top5.set_index('Country of Birth', inplace=True)
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.plot(kind='bar')

plt.title('Top 5 immigrants to NZ from 1949 to 2020')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```

```python
df_brazil = df.loc[df['Country of Birth']=='Brazil', years]

df_brazil= df_brazil[years].transpose()

df_brazil.plot(kind='bar')
plt.title('Brazilian immigrants to NZ from 1949 to 2020')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```
